```python
# ============================================================
# INDIAN POLITICAL SPEECHES - COMPLETE EDA NOTEBOOK
# ============================================================

# ============================================================
# INSTALL LIBRARIES
# ============================================================

!pip install pandas seaborn nltk wordcloud openpyxl -q

# ============================================================
# IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import nltk
import re

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.util import ngrams

from collections import Counter

from wordcloud import WordCloud

# ============================================================
# DOWNLOAD NLTK DATA
# ============================================================

nltk.download('punkt')
nltk.download('stopwords')

# ============================================================
# SEABORN STYLE
# ============================================================

sns.set_theme(style="whitegrid")

# ============================================================
# LOAD DATASET
# ============================================================

# CHANGE FILE NAME HERE
df = pd.read_csv('/content/cleaned_political_speeches.csv')

# ============================================================
# VIEW DATASET
# ============================================================

print(df.head())

print(df.info())

print(df.columns)

# ============================================================
# KEEP ONLY IMPORTANT COLUMNS
# ============================================================

df = df[
    ['speaker', 'year', 'text', 'words']
]

# ============================================================
# REMOVE NULL VALUES
# ============================================================

df = df.dropna(subset=['text'])

df = df.dropna(subset=['year'])

# ============================================================
# CONVERT YEAR TO INTEGER
# ============================================================

df['year'] = df['year'].astype(int)

# ============================================================
# REMOVE DUPLICATE SPEECHES
# ============================================================

df = df.drop_duplicates(subset=['text'])

# ============================================================
# TEXT CLEANING
# ============================================================

stop_words = set(stopwords.words('english'))

# CUSTOM POLITICAL STOPWORDS
custom_words = {
    'friends',
    'brothers',
    'sisters',
    'country',
    'people',
    'india',
    'today'
}

stop_words.update(custom_words)

# ============================================================
# CLEAN TEXT FUNCTION
# ============================================================

def clean_text(text):

    text = str(text).lower()

    # remove urls
    text = re.sub(r"http\S+", "", text)

    # remove punctuation/numbers
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text)

    return text.strip()

# ============================================================
# APPLY CLEANING
# ============================================================

df['clean_text'] = df['text'].apply(clean_text)

# ============================================================
# DATASET OVERVIEW
# ============================================================

print("Total speeches:", len(df))

print("Total speakers:", df['speaker'].nunique())

print("Years covered:",
      df['year'].min(),
      "-",
      df['year'].max())

# ============================================================
# SPEAKER DISTRIBUTION
# ============================================================

speaker_counts = df['speaker'].value_counts()

major_speakers = speaker_counts[speaker_counts > 20]

others_count = speaker_counts[speaker_counts <= 20].sum()

plot_data = major_speakers.copy()

plot_data['Others'] = others_count

plot_df = pd.DataFrame({
    'speaker': plot_data.index,
    'count': plot_data.values
})

plt.figure(figsize=(14,6))

sns.barplot(
    data=plot_df,
    x='speaker',
    y='count'
)

plt.xticks(rotation=45)

plt.title(
    "Distribution of Political Speeches by Speaker",
    fontsize=16
)

plt.xlabel("Speaker")

plt.ylabel("Number of Speeches")

plt.show()

# ============================================================
# TOP 15 SPEAKERS
# ============================================================

top_speakers = (
    df['speaker']
    .value_counts()
    .head(15)
    .reset_index()
)

top_speakers.columns = ['speaker', 'count']

plt.figure(figsize=(14,6))

sns.barplot(
    data=top_speakers,
    x='speaker',
    y='count'
)

plt.xticks(rotation=45)

plt.title(
    "Top 15 Speakers",
    fontsize=16
)

plt.xlabel("Speaker")

plt.ylabel("Number of Speeches")

plt.show()

# ============================================================
# YEAR-WISE ANALYSIS
# ============================================================

year_counts = (
    df['year']
    .value_counts()
    .sort_index()
    .reset_index()
)

year_counts.columns = ['year', 'count']

plt.figure(figsize=(14,6))

sns.barplot(
    data=year_counts,
    x='year',
    y='count'
)

plt.xticks(rotation=45)

plt.title(
    "Number of Political Speeches Per Year",
    fontsize=16
)

plt.xlabel("Year")

plt.ylabel("Speech Count")

plt.show()

# ============================================================
# YEAR-WISE LINEPLOT
# ============================================================

plt.figure(figsize=(14,6))

sns.lineplot(
    data=year_counts,
    x='year',
    y='count',
    marker='o'
)

plt.title(
    "Political Speech Trend Over Years",
    fontsize=16
)

plt.xlabel("Year")

plt.ylabel("Speech Count")

plt.show()

# ============================================================
# SPEECH LENGTH DISTRIBUTION
# ============================================================

plt.figure(figsize=(14,6))

sns.histplot(
    data=df,
    x='words',
    bins=30,
    kde=True
)

plt.title(
    "Speech Length Distribution",
    fontsize=16
)

plt.xlabel("Number of Words")

plt.ylabel("Frequency")

plt.show()

# ============================================================
# BOXPLOT
# ============================================================

plt.figure(figsize=(14,4))

sns.boxplot(
    x=df['words']
)

plt.title(
    "Speech Length Boxplot",
    fontsize=16
)

plt.show()

# ============================================================
# WORD CLOUD
# ============================================================

all_text = " ".join(df['clean_text'])

wordcloud = WordCloud(
    width=1200,
    height=600,
    background_color='white',
    stopwords=stop_words
).generate(all_text)

plt.figure(figsize=(15,8))

plt.imshow(wordcloud)

plt.axis('off')

plt.title(
    "Political Speech Word Cloud",
    fontsize=18
)

plt.show()

# ============================================================
# TOKENIZATION
# ============================================================

tokens = []

for text in df['clean_text']:

    words = word_tokenize(text)

    words = [
        word for word in words
        if word not in stop_words
        and len(word) > 2
    ]

    tokens.extend(words)

# ============================================================
# BIGRAM ANALYSIS
# ============================================================

bigrams_list = list(ngrams(tokens, 2))

bigram_freq = Counter(bigrams_list)

top_bigrams = bigram_freq.most_common(20)

bigram_df = pd.DataFrame(
    top_bigrams,
    columns=['bigram', 'frequency']
)

bigram_df['bigram'] = bigram_df['bigram'].apply(
    lambda x: ' '.join(x)
)

plt.figure(figsize=(14,8))

sns.barplot(
    data=bigram_df,
    x='frequency',
    y='bigram'
)

plt.title(
    "Top 20 Political Bigrams",
    fontsize=16
)

plt.xlabel("Frequency")

plt.ylabel("Bigram")

plt.show()

# ============================================================
# TRIGRAM ANALYSIS
# ============================================================

trigrams_list = list(ngrams(tokens, 3))

trigram_freq = Counter(trigrams_list)

top_trigrams = trigram_freq.most_common(20)

trigram_df = pd.DataFrame(
    top_trigrams,
    columns=['trigram', 'frequency']
)

trigram_df['trigram'] = trigram_df['trigram'].apply(
    lambda x: ' '.join(x)
)

plt.figure(figsize=(14,8))

sns.barplot(
    data=trigram_df,
    x='frequency',
    y='trigram'
)

plt.title(
    "Top 20 Political Trigrams",
    fontsize=16
)

plt.xlabel("Frequency")

plt.ylabel("Trigram")

plt.show()

# ============================================================
# SPEAKER VS YEAR HEATMAP
# ============================================================

heatmap_data = pd.crosstab(
    df['speaker'],
    df['year']
)

top_speaker_names = (
    df['speaker']
    .value_counts()
    .head(10)
    .index
)

heatmap_data = heatmap_data.loc[top_speaker_names]

plt.figure(figsize=(14,8))

sns.heatmap(
    heatmap_data,
    cmap='Blues',
    annot=True,
    fmt='d'
)

plt.title(
    "Speaker vs Year Heatmap",
    fontsize=16
)

plt.xlabel("Year")

plt.ylabel("Speaker")

plt.show()

# ============================================================
# AVERAGE SPEECH LENGTH PER SPEAKER
# ============================================================

avg_length = (
    df.groupby('speaker')['words']
    .mean()
    .sort_values(ascending=False)
    .head(15)
    .reset_index()
)

plt.figure(figsize=(14,6))

sns.barplot(
    data=avg_length,
    x='speaker',
    y='words'
)

plt.xticks(rotation=45)

plt.title(
    "Average Speech Length Per Speaker",
    fontsize=16
)

plt.xlabel("Speaker")

plt.ylabel("Average Word Count")

plt.show()

# ============================================================
# SAVE CLEANED DATASET
# ============================================================

df.to_csv(
    '/content/cleaned_political_speeches.csv',
    index=False
)

print("Cleaned dataset saved successfully.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
import re

import nltk

nltk.download('punkt')
nltk.download('punkt_tab')


from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.util import ngrams
from collections import Counter

nltk.download('punkt')
nltk.download('stopwords')

# Load dataset
df = pd.read_csv('/content/cleaned_political_speeches.csv')

# Stopwords
stop_words = set(stopwords.words('english'))

# Clean text
def clean_text(text):

    text = str(text).lower()

    text = re.sub(r"[^a-zA-Z\s]", "", text)

    return text

df['clean_text'] = df['text'].apply(clean_text)

# Tokenization
tokens = []

for text in df['clean_text']:

    words = word_tokenize(text)

    words = [
        word for word in words
        if word not in stop_words
        and len(word) > 2
    ]

    tokens.extend(words)

# Generate bigrams
bigrams_list = list(ngrams(tokens, 2))

bigram_freq = Counter(bigrams_list)

top_bigrams = bigram_freq.most_common(20)

# Dataframe
bigram_df = pd.DataFrame(
    top_bigrams,
    columns=['bigram', 'frequency']
)

bigram_df['bigram'] = bigram_df['bigram'].apply(
    lambda x: ' '.join(x)
)

# Plot
plt.figure(figsize=(14,8))

sns.barplot(
    data=bigram_df,
    x='frequency',
    y='bigram'
)

plt.title(
    "Top 20 Political Bigrams",
    fontsize=16
)

plt.xlabel("Frequency")

plt.ylabel("Bigram")

# Save plot
plt.savefig(
    '/content/bigrams.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```
